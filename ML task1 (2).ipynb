```python
#SUPRIYA K 24BAD118
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data.csv", encoding="ISO-8859-1")
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())
df["Sales"] = df["Quantity"] * df["UnitPrice"]
sales_per_product = df.groupby("Description")["Sales"].sum().head(10)
sales_per_product.plot(kind="bar")
plt.show()
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
sales_over_time = df.groupby(df["InvoiceDate"].dt.date)["Sales"].sum()
plt.plot(sales_over_time)
plt.show()
```

```python
#SUPRIYA K 24BAD118
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("diabetes.csv")

print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())

cols_with_zero = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)

print(df.isnull().sum())

plt.hist(df["Glucose"].dropna(), bins=20)
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.title("Distribution of Glucose Levels")
plt.show()

plt.hist(df["Age"], bins=15)
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.title("Age Distribution of Patients")
plt.show()

plt.boxplot(df["Glucose"].dropna())
plt.ylabel("Glucose Level")
plt.title("Boxplot of Glucose Levels")
plt.show()

plt.boxplot(df["Age"])
plt.ylabel("Age")
plt.title("Boxplot of Age")
plt.show()

pattern_analysis = df.groupby("Outcome").mean()
print(pattern_analysis)
```

```python
#SUPRIYA K 24BAD118
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Housing.csv")

df.head()
df.info()
df.describe()

missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_values
missing_percentage
numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(df["area"], df["price"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Price vs Area")
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(df["bedrooms"], df["price"])
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.title("Price vs Bedrooms")
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(df["bathrooms"], df["price"])
plt.xlabel("Bathrooms")
plt.ylabel("Price")
plt.title("Price vs Bathrooms")
plt.show()
```

```python
#SUPRIYA K 24BAD118
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("marketing_campaign.csv", sep="\t")

df.head()
df.info()
df.describe()

missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_values
missing_percentage

df["Age"] = 2024 - df["Year_Birth"]

age_counts = df["Age"].value_counts().sort_index()

plt.figure(figsize=(8, 5))
plt.bar(age_counts.index, age_counts.values)
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.title("Customer Age Distribution")
plt.show()

plt.figure(figsize=(6, 4))
plt.boxplot(df["Income"].dropna())
plt.ylabel("Income")
plt.title("Customer Income Distribution")
plt.show()

spending_columns = [
    "MntWines",
    "MntFruits",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
    "MntGoldProds"
]

plt.figure(figsize=(10, 6))
df[spending_columns].boxplot()
plt.ylabel("Amount Spent")
plt.title("Customer Spending Patterns by Product Category")
plt.show()
```
